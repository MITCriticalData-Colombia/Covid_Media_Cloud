```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
main_path = 'Google_Mobility/'
out_dir = 'Mobility_Colombia/'
```

# Global Mobility

```python
# Source: https://www.google.com/covid19/mobility/
golbal_mobility_df = pd.read_csv(main_path + 'Global_Mobility_Report.csv')
golbal_mobility_df.drop(columns=['metro_area', 'census_fips_code', 'place_id', 'iso_3166_2_code'], inplace=True)
golbal_mobility_df
```

# Mobility in Colombia

```python
colombia_mobility_df = golbal_mobility_df[golbal_mobility_df['country_region_code'] == 'CO']
colombia_mobility_df
```

```python
# Departments
pd.unique(colombia_mobility_df['sub_region_1'])
```

```python
# Municipalities
pd.unique(colombia_mobility_df['sub_region_2'])
```

```python
# Null values
colombia_mobility_df.isna().sum()[colombia_mobility_df.isna().sum()>0].plot(kind='bar')
print(len(colombia_mobility_df))
colombia_mobility_df.isna().sum()
```

```python
# Use pandas.to_datetime() to convert string to datetime format
colombia_mobility_df['date'] = pd.to_datetime(colombia_mobility_df['date'])
colombia_mobility_df.dtypes
```

# Mobility by Department

```python
colombia_mobility_df = colombia_mobility_df[colombia_mobility_df['sub_region_1'].notna()]
colombia_mobility_df.isna().sum()
```

```python
mobility_by_department = colombia_mobility_df.groupby([colombia_mobility_df['date'].dt.date,'sub_region_1']).mean()
mobility_by_department
```

```python
position_index = mobility_by_department.index.names.index('date')
mobility_by_department = mobility_by_department.unstack(position_index)
mobility_by_department
```

### Save raw data by department

```python
mobility_by_department.stack(1).to_csv(os.path.join(main_path, out_dir, 'mobility_by_department_raw.csv'),index=True)
```

## Data cleaning

```python
## Null values for department
number_dates = mobility_by_department.T.shape[0]

# Threshold
threshold = number_dates/4

# Just keep Municipalities with null values less than total values / 4
mobility_by_department = mobility_by_department.T.loc[:,mobility_by_department.T.isna().sum() < threshold]

mobility_by_department = mobility_by_department.fillna(mobility_by_department.mean())
mobility_by_department.isna().sum()
```

```python
mobility_by_department = mobility_by_department.T
mobility_by_department
```

```python
mobility_by_department = mobility_by_department.stack(1)
```

```python
mobility_by_department.to_csv(os.path.join(main_path, out_dir, 'mobility_by_department.csv'),index=True)
```

# By country

```python
colombia_mobility = colombia_mobility_df.groupby([colombia_mobility_df['date'].dt.date]).mean()
colombia_mobility
```

```python
colombia_mobility.isnull().sum()
```

```python
colombia_mobility.to_csv(os.path.join(main_path, out_dir, 'mobility_colombia.csv'),index=True)
```

```python
# Covid cases in time
plt.figure(figsize = (12, 8))
plt.plot(colombia_mobility.index, colombia_mobility['retail_and_recreation_percent_change_from_baseline'], color ='blue', label='retail_and_recreation_percent_change_from_baseline')
plt.plot(colombia_mobility.index, colombia_mobility['grocery_and_pharmacy_percent_change_from_baseline'], color ='red', label='grocery_and_pharmacy_percent_change_from_baseline')
plt.plot(colombia_mobility.index, colombia_mobility['workplaces_percent_change_from_baseline'], color ='green', label='workplaces_percent_change_from_baseline')
plt.plot(colombia_mobility.index, colombia_mobility['residential_percent_change_from_baseline'], color ='black', label='residential_percent_change_from_baseline')
plt.xlabel('date of mobility score in Colombia') 
plt.ylabel('Mobility score')
plt.legend()
```
