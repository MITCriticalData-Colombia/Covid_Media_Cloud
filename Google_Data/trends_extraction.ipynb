# Connecting to Google

```python
import pandas as pd                 
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
# Source: multiTimeline.csv
col_trends = pd.read_csv('Google_Trends/multiTimeline.csv', skiprows=1)
col_trends['Semana'] = pd.to_datetime(col_trends['Semana'])
col_trends
```

```python
# Covid cases in time
plt.figure(figsize = (12, 8))
plt.plot(col_trends.Semana, col_trends['tos: (Colombia)'], color ='blue', label='cough')
plt.plot(col_trends.Semana, col_trends['fiebre: (Colombia)'], color ='red', label='fever')
plt.plot(col_trends.Semana, col_trends['cansado: (Colombia)'], color ='green', label='tired')
plt.plot(col_trends.Semana, col_trends['diarrea: (Colombia)'], color ='black', label='diarrhea')
plt.plot(col_trends.Semana, col_trends['dolor de garganta: (Colombia)'], color ='purple', label='throat pain')

plt.xlabel('Week') 
plt.ylabel('Trend score')
plt.legend()
plt.show()
```
