```python
import pandas as pd                 
import matplotlib.pyplot as plt
```

## Read file

```python
# Function to read COVID-19 symptoms file
def read_file(Country):
    df_trends = pd.read_csv(f'Google_Trends (COVID-19 Symptoms)/{Country}.csv', skiprows=1)
    df_trends['Week'] = pd.to_datetime(df_trends['Week'])
    return df_trends
```

## Plot dataframe

```python
# Function to plot COVID-19 symptoms in time
def plot_symptoms_df(df, country):
    plt.figure(figsize = (12, 8))
    plt.plot(df.Week, df[f'Cough: ({country})'], color ='blue', label=f'cough in : {country}')
    plt.plot(df.Week, df[f'Fever: ({country})'], color ='red', label=f'fever in : {country}')
    plt.plot(df.Week, df[f'Headache: ({country})'], color ='green', label=f'Headache in : {country}')

    plt.xlabel('Week') 
    plt.ylabel('Trend score')
    plt.legend()
    plt.show()
```

# Examples

### Example 1:

```python
# Select Country
COUNTRY = 'Colombia'
# Read file
df_trends = read_file(COUNTRY)
df_trends
```

```python
# Plot data of Colombia
plot_symptoms_df(df=df_trends, country=COUNTRY)
```

```python
# Select Country
COUNTRY = 'Argentina'
# Read file
df_trends = read_file(COUNTRY)
df_trends
```
